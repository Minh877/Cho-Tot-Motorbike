## Pyplot API vs Object-Oriented (OO) API
* Pyplot API --> Quickly
* OO API --> Advanced

### Pyplot API

```python
x = np.linspace(0, 10, 1000)
x
```

```python
plt.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x), color = 'blue', linestyle = "dashed", label = 'Sin(x)');
plt.plot(x, np.cos(x), color = 'red', label = 'Cos(x)');

plt.title('Sin(x) and Cos(x)')
plt.xlabel('X')
plt.ylabel('f(x)')

# plt.xlim([0,4])
# plt.ylim([-0.75,0.75])
# plt.axis([0, 4, -0.75, 0.75])
# plt.axis('tight')
plt.axis('equal')

plt.legend();
```

### Object-Oriented API

```python
#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plots
fig, ax = plt.subplots(figsize = (5,5)) #Figure size = Width & Height of the plot

#3. Plot the data
ax.plot(x, y);

ax.set(title = "A simple plot"
      , xlabel = "x-axis"
      , ylabel = "y-axis");
```

## 2. Most common types of Matplotlib plots
* line
* scatter
* bar
* hist
* subplot()

### Line

```python
# Create an array
x = np.linspace(0, 10, 100)
x[:5]
```

```python
# OO API plot a line chart
fig, ax = plt.subplots()
ax.plot(x, x**3);
```

### Scatter

```python
# Pyplot API
plt.scatter(x, np.exp(x)); # y = e^(x)
```

```python
# OO API

# Prepare new data
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000*rng.rand(100)

fig, ax = plt.subplots()

img1 = ax.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.3);

fig.colorbar(img1);
```

### Bar
* Vertical
* Horizontal

```python
# Prepare data from Python dictionary
soft_drink_prices = {"Coke": 10
                    , "Pepsi": 12
                    , "Sprite": 8}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax.set(title="Bach hoa xanh's soft drink prices", ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()));
```

### Histogram

```python
# Prepare data
np.random.seed(42)
students_height = np.random.normal(170, 10, 250)
students_height[:10]
```

```python
# Pyplot API
plt.hist(students_height);
```

```python
plt.hist(students_height, bins=30);
```

### Subplots
* Multiple plots on one figure

```python
# Option 1:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))

#Plot data to each axis
ax1.plot(x, x/2);
ax1.plot(x, x*2);

ax2.scatter(np.random.random(10), np.random.random(10));

ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());

ax4.hist(students_height, bins=30);
```

```python
# Option 2:
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))

#Plot data to each axis
ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x*2);

ax[0,1].scatter(np.random.random(10), np.random.random(10));

ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values());

ax[1,1].hist(students_height, bins=30);
```
